```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_train = pd.read_csv('/kaggle/input/flipkart-product-dataset/train.csv') 
df_test = pd.read_csv('/kaggle/input/flipkart-product-dataset/test.csv')
```

Removing null values from train data

```python
df_train = df_train.dropna()
```

Removing '%' from offer

```python
df_train['Offer %']=df_train['Offer %'].str.rstrip('%')
df_train['Offer %']
```

```python
dummy = pd.get_dummies(df_train[['maincateg','platform']],drop_first=True)
df_train=pd.concat([df_train,dummy],axis=1)
```

offer , price1 are not present in test data so drop them. Id, tiitle, platform not needed

```python
df_train=df_train.drop(['id','title','platform','maincateg','price1','Offer %'],axis=1)
```

Clean test data

```python
df_test= df_test.dropna()
df_test
```

```python
dummy1 = pd.get_dummies(df_test[['maincateg','platform']],drop_first=True)
df_test = pd.concat([df_test,dummy1],axis=1)
df_test = df_test.drop(['maincateg','platform','id','title'],axis=1)
```

Predict the rating using deep learning

```python
X_train=df_train.drop('Rating',axis=1).values
y_train = df_train['Rating']
X_test = df_test.drop('Rating',axis=1).values
y_test = df_test['Rating']
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(X_train,y_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()
model.add(Dense(10,activation='relu'))
model.add(Dense(10,activation='relu'))
model.add(Dense(10,activation='relu'))
model.add(Dense(10,activation='relu'))

model.add(Dense(1))


model.compile(optimizer='rmsprop',loss='mse')

```

```python
model.fit(X_train,y_train,validation_data=(X_test,y_test),epochs=300)
```

```python
df_loss = pd.DataFrame(model.history.history)
df_loss.plot()
```

```python
pred = model.predict(X_test)
```

Now lets predicts single value say rating of index 1

```python
df_test['Rating'][1]
```

```python
single_data = df_test.drop('Rating',axis=1).iloc[1]
single_data.values.reshape(-1,11)
scaled_single_data = scaler.transform(single_data.values.reshape(-1,11))
pred1 = model.predict(scaled_single_data)
print()
print('predict_first:')
print(pred1)
```
